data preparation for modeling and machine learning technique

```python
"""
The logarithmic transformation  for handling strong asymmetries. The logarithmic transformation 
reduces the distance between extreme values and small values while keeping the relationships proportional.    
"""
for var in quantitatives_variables:
    courses_data[var] = np.log1p(courses_data[var])
```

```python
#True = 1 , False = 0
courses_data['is_paid'] = courses_data['is_paid'].astype(int)
```

```python
level_mapping = {
    'Beginner Level': 0,
    'Intermediate Level': 1,
    'All Levels': 2,
    'Expert Level': 3
}

courses_data['level'] = courses_data['level'].map(level_mapping)
```

```python

from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
courses_data['subject'] = label_encoder.fit_transform(courses_data['subject'])
```
